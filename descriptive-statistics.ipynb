```python
import pandas, numpy
from matplotlib import pyplot
import scipy.stats
from collections import Counter
from ipywidgets import interact, interactive, fixed, interact_manual
from sklearn.linear_model import LinearRegression, LogisticRegression
```

```python
teamDict = {
    0: "Angels",
    1: "Astros",
    2: "Athletics",
    3: "Blue Jays",
    4: "Braves",
    5: "Brewers",
    6: "Cardinals",
    7: "Cubs",
    8: "D-Backs",
    9: "Dodgers",
    10: "Giants",
    11: "Indians",
    12: "Mariners",
    13: "Marlins",
    14: "Mets",
    15: "Nationals",
    16: "Orioles",
    17: "Padres",
    18: "Phillies",
    19: "Pirates",
    20: "Rangers",
    21: "Rays",
    22: "Reds",
    23: "Red Sox",
    24: "Rockies",
    25: "Royals",
    26: "Tigers",
    27: "Twins",
    28: "White Sox",
    29: "Yankees"
}
```

```python
baseball_stats = pandas.read_csv("merged_data_final.csv")
```

```python
# Find average runs per game for each team
runs_avg = []
runs_var = []

runsAvgDict = {}
runsVarDict = {}

for i in range(len(teamDict)):
    team_runs_list = []
    
    for j in baseball_stats.values:
        if(j[10] == teamDict[i]):
            team_runs_list.append(j[13])
        elif(j[3] == teamDict[i]):
            team_runs_list.append(j[6])
    
    team_runs_list = numpy.array(team_runs_list)
    runs_avg.append(team_runs_list.mean())
    runs_var.append(team_runs_list.var())
    
    runsAvgDict[teamDict[i]] = runs_avg[i]
    runsVarDict[teamDict[i]] = runs_var[i]

pyplot.bar(runsAvgDict.keys(), runsAvgDict.values())
pyplot.xticks(rotation='vertical')
pyplot.title("Average Runs per Game")
pyplot.show()

pyplot.bar(runsVarDict.keys(), runsVarDict.values())
pyplot.xticks(rotation='vertical')
pyplot.title("Variance of Runs per Game")
pyplot.show()

mlb_runs_avg = baseball_stats['total_runs'].mean() / 2
print("MLB Average Runs: ", mlb_runs_avg)
```

```python
# Find average hits per game for each team

hits_avg = []
hits_var = []

hitsAvgDict = {}
hitsVarDict = {}

for i in range(len(teamDict)):
    team_hits_list = []
    
    for j in baseball_stats.values:
        if(j[10] == teamDict[i]):
            team_hits_list.append(j[12])
        elif(j[3] == teamDict[i]):
            team_hits_list.append(j[5])
    
    team_hits_list = numpy.array(team_hits_list)
    hits_avg.append(team_hits_list.mean())
    hits_var.append(team_hits_list.var())
    
    hitsAvgDict[teamDict[i]] = hits_avg[i]
    hitsVarDict[teamDict[i]] = hits_var[i]

pyplot.bar(hitsAvgDict.keys(), hitsAvgDict.values())
pyplot.xticks(rotation='vertical')
pyplot.title("Average Hits per Game")
pyplot.show()

pyplot.bar(hitsVarDict.keys(), hitsVarDict.values())
pyplot.xticks(rotation='vertical')
pyplot.title("Variance of Hits per Game")
pyplot.show()

mlb_hits_avg = (baseball_stats['away_team_hits'] + baseball_stats['home_team_hits']).mean() / 2
print("MLB Average Hits: ", mlb_hits_avg)
```

```python
# Find average home game temperature

temp_avg = []
temp_var = []

tempAvgDict = {}
tempVarDict = {}

for i in range(len(teamDict)):
    team_temp_list = []
    
    for j in baseball_stats.values:
        if(j[10] == teamDict[i]):
            team_temp_list.append(j[17])
            
    team_temp_list = numpy.array(team_temp_list)
    temp_avg.append(team_temp_list.mean())
    temp_var.append(team_temp_list.var())
    
    tempAvgDict[teamDict[i]] = temp_avg[i]
    tempVarDict[teamDict[i]] = temp_var[i]

pyplot.bar(tempAvgDict.keys(), tempAvgDict.values())
pyplot.xticks(rotation='vertical')
pyplot.title("Average Temperature at Home Games (Fahrenheit)")
pyplot.show()

pyplot.bar(tempVarDict.keys(), tempVarDict.values())
pyplot.xticks(rotation='vertical')
pyplot.title("Variance of Temperature at Home Games (Fahrenheit)")
pyplot.show()

mlb_average_temperature = baseball_stats['temperature'].mean()
print("MLB Average Temperature: ", mlb_average_temperature)
```

```python
# Find average home game wind speed

wind_avg = []
wind_var = []

windAvgDict = {}
windVarDict = {}

for i in range(len(teamDict)):
    team_wind_list = []
    
    for j in baseball_stats.values:
        if(j[10] == teamDict[i]):
            team_wind_list.append(j[18])
            
    team_wind_list = numpy.array(team_wind_list)
    wind_avg.append(team_wind_list.mean())
    wind_var.append(team_wind_list.var())
    
    windAvgDict[teamDict[i]] = wind_avg[i]
    windVarDict[teamDict[i]] = wind_var[i]

pyplot.bar(windAvgDict.keys(), windAvgDict.values())
pyplot.xticks(rotation='vertical')
pyplot.title("Average Wind Speed at Home Games (MPH)")
pyplot.show()

pyplot.bar(windVarDict.keys(), windVarDict.values())
pyplot.xticks(rotation='vertical')
pyplot.title("Variance of Wind Speed at Home Games (MPH)")
pyplot.show()

mlb_avg_wind = baseball_stats['wind_speed'].mean()
print("MLB Average Wind Speed: ", mlb_avg_wind)
```

```python
# Find average exit velocity for each team

ev_avg = []
ev_var = []

evAvgDict = {}
evVarDict = {}

for i in range(len(teamDict)):
    team_ev_list = []
    
    for j in baseball_stats.values:
        if(j[10] == teamDict[i]):
            team_ev_list.append(j[32])
        elif(j[3] == teamDict[i]):
            team_ev_list.append(j[43])
    
    team_ev_list = numpy.array(team_ev_list)
    ev_avg.append(team_ev_list.mean())
    ev_var.append(team_ev_list.var())
    
    evAvgDict[teamDict[i]] = ev_avg[i]
    evVarDict[teamDict[i]] = ev_var[i]

pyplot.bar(evAvgDict.keys(), evAvgDict.values())
pyplot.xticks(rotation='vertical')
pyplot.title("Average Exit Velocity (MPH)")
pyplot.ylim(bottom=80, top=86)
pyplot.show()

pyplot.bar(evVarDict.keys(), evVarDict.values())
pyplot.xticks(rotation='vertical')
pyplot.title("Variance of Exit Velocity (MPH)")
pyplot.show()

mlb_avg_ev = (baseball_stats['Home_Avg EV'] + baseball_stats['Away_Avg EV']).mean() / 2
print("MLB Avg. Exit Velocity: ", mlb_avg_ev)
```
